# Let's learn about birds


```python
import pandas as pd
import matplotlib.pyplot as plt

birds = pd.read_csv('../../data/birds.csv')

birds.head()
```

```python
plt.title('Max Wingspan in cm')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.xticks(rotation = 45)

x = birds['Name']
y = birds['MaxWingspan']

plt.plot(x,y)

plt.show()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, 'bo')
    if birds['MaxWingspan'][i] > 500:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=12)
    
plt.show()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']:
        plt.plot(x, y, 'bo')
        
plt.show()
```

```python
birds.plot(x='Category', kind='bar',stacked=True,title='Birds of Minnesota')
```

```python
category = birds.value_counts(birds['Category'], sort=True)
plt.rcParams['figure.figsize'] = [6,12]
category.plot.barh()
```

```python
max_length = birds['MaxLength']
category = birds['Category']

plt.barh(category,max_length)

plt.show()
```

```python
max_length = birds['MaxLength']
min_length = birds['MinLength']
category = birds['Category']

plt.barh(category,max_length)
plt.barh(category,min_length)

plt.show()
```
